## How to label customers whose aggregated `amount_spent` is in a given quantile range?


We break down the main problem into several sub problems:

#### Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?

#### Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?

```python
# your code here
customer_spending = orders.groupby('CustomerID')['amount_spent'].sum().reset_index()

# Print the first few rows to see the aggregated data
customer_spending.head()
```

```python
percentiles = orders['amount_spent']

# Calculate the 95th percentile
percentile_95 = np.percentile(percentiles, 95)

# Calculate the 75th percentile
percentile_75 = np.percentile(percentiles, 75)

print(f"95th percentile: {percentile_95}")
print(f"75th percentile: {percentile_75}")
```

```python
top_customers = pd.DataFrame(orders, columns=['CustomerID', 'amount_spent'])

def best_customers(row, percentile_75, percentile_95):
    if row['amount_spent'] > percentile_95:
        return 'VIP'
    elif percentile_75 <= row['amount_spent'] <= percentile_95:
        return 'Preferred customer'
    else:
        return 'Regular customer'
    
top_customers['TopCustomer'] = top_customers.apply(best_customers, args=(percentile_75, percentile_95), axis=1)

top_customers.head(25)
```

```python
custom_sort_order = {'VIP': 0, 'Preferred customer': 1, 'Regular customer': 2}

# Map 'TopCustomer' to the custom sorting order
top_customers['TopCustomer'] = top_customers['TopCustomer'].map(custom_sort_order)

# Invert the custom_sort_order dictionary to map back to labels
inverse_sort_order = {v: k for k, v in custom_sort_order.items()}
top_customers['TopCustomer'] = top_customers['TopCustomer'].map(inverse_sort_order)

# Sort 'TopCustomer' in descending order
top_customers = top_customers.sort_values(by='TopCustomer', ascending=False)

# Reset the index after sorting (optional)
top_customers = top_customers.reset_index(drop=True)

# Now, the 'TopCustomer' column contains labels (VIP, Preferred customer, Regular customer)

top_customers.head(50)
```

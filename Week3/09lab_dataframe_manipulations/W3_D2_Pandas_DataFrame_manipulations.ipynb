```python
import pandas as pd

superstore_df = pd.read_csv('Sample - EU Superstore.csv')
superstore_df.head() #or display(superstore)

```

```python
superstore_df['Profitability ratio']= superstore_df['Profit']/superstore_df['Sales']
superstore_df.drop(columns=['Profitability ration'], inplace=True)
superstore_df.head()
```

```python
# For complex tasks create a function
#E.g I want to know for each row if the sale was profitable or not

superstore_df['Profit']>0

#But I want to be more granualar and to know if the profit is lass than 0 = not profitable
#BUT if > 60 box, it is okayish profitable
#So we want the result to be granualar (not profiable, okayish profitable...)
```

```python
def profitable_advanced(profit_item):
  if profit_item < 0:
    return 'not profitable'
  elif profit_item < 100:
    return 'somewhat profitable'
  else:
    return 'very profitable'
```

```python
#Now we wanna make the function works on every row of the dataset.
# For this, we are going to use apply
# 1. We take the column of the function we wanna use the .apply and we put the name of the function in ()
# 2. Create new variable to add it to the main dataset

superstore_df['profit_flag'] = superstore_df['Profit'].apply(profitable_advanced)
display(superstore_df)
```

```python
# Then we wanna figure out if orders have been delayed or not
# If ship date is more than 5 days after the order date = shipment is late

#1. We need to convert string date into date objects

from datetime import datetime #we import a subpackage from the package datetime
from datetime import timedelta #timedelta = calculates the difference between two specific times
#You get the delta when you substract/or add two values. Eg date reference = today
#timedelta = today - day in 7 days ago OR today + day in 7 days
```

```python
datetime.strptime('06/04/23', '%d/%m/%y')
#strp = it takes a string and returns it as an object date/time
# We use two arguments : thing we want to take the time from + the format in which it is given to us
#if %Y = will display 4 numbers / %y = only 2 numbers
```

```python
#To calculate the delta between order date and shipping date we do :

#1. I convert date string into date object. I create a function for that
def transform_string_to_dateobject(date_str):
  return datetime.strptime(date_str, '%d/%m/%Y')
```

```python
superstore_df['Order Date'] = superstore_df['Order Date'].apply(transform_string_to_dateobject) #apply the function created to the dataframe
#We overwrite the original date to have date object
superstore_df['Ship Date'] = superstore_df['Ship Date'].apply(transform_string_to_dateobject)
```

```python
superstore_df.head(3)
```

```python
#We apply the function to all row with axis = 1
def delayed_shipments(row):
  delta = row['Ship Date']-row['Order Date'] #the result is the delta / We take later date-original date
  if delta > timedelta(5): #very important to add time delta here
    return 'Delayed'
  else:
    return 'On time'
```

```python
#Dataframe is a two dimensional object (column and row). Here we want to apply the function row by row, so I add axis=1
superstore_df['Delayed ?'] = superstore_df.apply(delayed_shipments, axis = 1)
superstore_df.sample(3)
```

```python
#Check the entries we have
superstore_df.apply(delayed_shipments, axis = 1).value_counts()
```

```python
# Categorise clients into two categories : Priority 1 and Priority 2
# Priority 1 : if Segment = 'Corporate' and Ship Mode = 'First or Second class'
#The ship mod is First Class
# Priority 2 otherwise

#My DIY
def orders_priority(row):
  if (row['Segment'] == 'Corporate') and (row['Ship Mode'] == 'First class' or row['Ship Mode'] == 'Second class'): #we want to apply it to the whole row!
    return 'Priority 1'
  elif row['Ship Mode'] == 'First Class': #Applies to the 3 segments
    return 'Priority 1'
  else :
    return 'Priority 2'
```

```python
superstore_df.apply(orders_priority, axis = 1).value_counts()
```
